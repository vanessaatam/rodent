```python
import pandas as pd

# Load the dataset
file_path = 'DSNY_Monthly_Tonnage_Data_2019-2024.csv'  
data = pd.read_csv(file_path)

# Convert 'MONTH' column to datetime to filter by year
data['MONTH'] = pd.to_datetime(data['MONTH'], errors='coerce', format='%Y / %m')

# Filter for rows where the year is 2023
data_2023 = data[data['MONTH'].dt.year == 2023]

# Select relevant columns
columns_of_interest = ['MONTH','BOROUGH', 'BOROUGH_ID', 'REFUSETONSCOLLECTED', 'PAPERTONSCOLLECTED', 'MGPTONSCOLLECTED', 'RESORGANICSTONS']
filtered_data_2023 = data_2023[columns_of_interest]

# Save to a new CSV file (optional)
output_file_path = 'Filtered_Tonnage_Data_2023.csv'
filtered_data_2023.to_csv(output_file_path, index=False)

# Display the processed data
filtered_data_2023.head()
```
